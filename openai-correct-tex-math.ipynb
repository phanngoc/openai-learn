```python
import openai
from openai import OpenAI
from dotenv import load_dotenv
import os
load_dotenv()

apiKey = os.getenv("OPENAI_API_KEY")
client = OpenAI(api_key=apiKey)
print(apiKey)
```

```python
def ask_chat(text, system_prompt = ''):
    messages = [
        {"role": "system", "content": f"{system_prompt}"},
        {"role": "user", "content": text},
    ]
    response = client.chat.completions.create(
        model='gpt-3.5-turbo',
        messages=messages,
        max_tokens=1000,
        top_p=1.0,
        n=1,
        stop=None,
        temperature=0.7)

    summary = response.choices

    return summary
```

```python
with open('input.txt', 'r') as file:
    content = file.read()

print(content)

t = ask_chat(content, system_prompt='Correct font teX math equations, convert it to pure TeX format')
```
